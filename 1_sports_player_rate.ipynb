```python
import pandas as pd
import numpy as np

# 지수 평활 함수
def exponential_smoothing(data, alpha):
    smoothed = np.zeros_like(data, dtype=float)
    smoothed[0] = data[0]  # 초기 값은 첫 번째 관측값
    for t in range(1, len(data)):
        smoothed[t] = alpha * data[t] + (1 - alpha) * smoothed[t - 1]
    return smoothed
```

```python
import numpy as np
import pandas as pd

def exponential_smoothing(data, alpha):
    smoothed = np.zeros_like(data, dtype=float)
    smoothed[0] = data[0]  # 초기 값은 첫 번째 관측값
    for t in range(1, len(data)):
        smoothed[t] = alpha * data[t] + (1 - alpha) * smoothed[t - 1]
    return smoothed

# 선수별 지수 평활 적용
for name in player_list:
    player_rate_unweighted = np.empty([1, 4])  # [player, time, result, rate]
    time_weight = 1
    rates = []

    # rate_raw_arr에서 해당 선수의 데이터를 수집
    for i in range(len(rate_raw_arr)):
        if rate_raw_arr[i, 0] == name:
            rates.append(rate_raw_arr[i, 3])  # 'rate'가 세 번째 열에 위치한다고 가정

    # 지수 평활 적용
    alpha = 0.7
    smoothed_rates = exponential_smoothing(np.array(rates), alpha)

    # 평활 결과 저장
    for j, rate in enumerate(smoothed_rates):
        arr = np.array([name, time_weight, rate_raw_arr[j, 2], rate])  # time_weight와 result 포함
        player_rate_unweighted = np.vstack((player_rate_unweighted, arr))
        time_weight += 1

    # 첫 번째 빈 행 제거
    player_rate_unweighted = player_rate_unweighted[1:]

    # 데이터프레임 생성 및 저장
    player_rate_unweighted_df = pd.DataFrame(player_rate_unweighted, columns=["player", "time", "result", "rate"])
    player_rate_unweighted_df.to_excel(f"data/rate_{name}.xlsx", index=False)
```

```python
import numpy as np
import pandas as pd

weighted_player_rate = np.empty([1, 4])
epsilon = 1e-5  # 안정성을 위한 작은 값

for name in player_list:
    file = "data/rate_" + name + ".xlsx"
    data = pd.read_excel(file)

    # 접전 경기에 더 높은 가중치를 부여
    data["inverse_result"] = 1 / (data["result"] + epsilon)

    # 총 가중치 계산
    sum_weight_time = sum(data["time"])
    if sum_weight_time != 0:
        # 가중치 적용 데이터 계산
        data["time*rate"] = data["rate"] * data["time"]
        data["inverse_result*rate"] = data["rate"] * data["inverse_result"]

        # 분산의 역수를 사용한 가중치 계산
        weight_time = 1 / (data["time*rate"].var() + epsilon)
        weight_inverse_result = 1 / (data["inverse_result*rate"].var() + epsilon)

        # 정규화
        total_weight = weight_time + weight_inverse_result
        weight_time /= total_weight
        weight_inverse_result /= total_weight

        # 최종 계산
        rate_time = sum(data["time*rate"]) / sum_weight_time
        rate_inverse_result = sum(data["inverse_result*rate"]) / sum_weight_time
        overall = (weight_time * rate_time) + (weight_inverse_result * rate_inverse_result)

        # 결과 저장
        arr = np.array([name, rate_time, rate_inverse_result, overall])
        weighted_player_rate = np.vstack((weighted_player_rate, arr))

# DataFrame 생성 및 저장
weighted_player_rate = weighted_player_rate[1:]
df = pd.DataFrame(weighted_player_rate, columns=["player", "rate_time", "rate_inverse_result", "overall"])
df = df.round(3)  # 소수점 3자리로 반올림
df.to_excel("./player_rates_weighted_avg.xlsx", index=False)
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

weighted_player_rate = []

epsilon = 1e-5  # 안정성을 위한 작은 값

# 모든 선수의 값을 저장할 리스트
all_data = []

for name in player_list:
    file = "data/rate_" + name + ".xlsx"
    data = pd.read_excel(file)

    # 접전 경기에 더 높은 가중치를 부여
    data["inverse_result"] = 1 / (data["result"] + epsilon)

    # 총 가중치 계산
    sum_weight_time = sum(data["time"])
    if sum_weight_time != 0:
        # 가중치 적용 데이터 계산
        data["time*rate"] = data["rate"] * data["time"]
        data["inverse_result*rate"] = data["rate"] * data["inverse_result"]

        # rate_time과 rate_inverse_result 계산
        rate_time = sum(data["time*rate"]) / sum_weight_time
        rate_inverse_result = sum(data["inverse_result*rate"]) / sum_weight_time

        # 선수 데이터 저장
        all_data.append([name, rate_time, rate_inverse_result])

# DataFrame 생성
df = pd.DataFrame(all_data, columns=["player", "rate_time", "rate_inverse_result"])

# 표준화
scaler = StandardScaler()
df[["rate_time_std", "rate_inverse_result_std"]] = scaler.fit_transform(df[["rate_time", "rate_inverse_result"]])

# 가중치 및 최종 계산
for _, row in df.iterrows():
    rate_time_std = row["rate_time_std"]
    rate_inverse_result_std = row["rate_inverse_result_std"]

    # 가중치 계산
    weight_time = 1 / (data["time*rate"].var() + epsilon)
    weight_inverse_result = 1 / (data["inverse_result*rate"].var() + epsilon)

    # 정규화
    total_weight = weight_time + weight_inverse_result
    weight_time /= total_weight
    weight_inverse_result /= total_weight

    # 최종 점수 계산
    overall = (weight_time * rate_time_std) + (weight_inverse_result * rate_inverse_result_std)
    weighted_player_rate.append([row["player"], rate_time_std, rate_inverse_result_std, overall])

# DataFrame 생성 및 저장
final_df = pd.DataFrame(weighted_player_rate, columns=["player", "rate_time_std", "rate_inverse_result_std", "overall"])
final_df = final_df.round(3)  # 소수점 3자리로 반올림
# final_df.to_excel("./player_rates_weighted_avg_standardized.xlsx", index=False)
```

## 부상예측 확률로 가중치 부여하기
- 직관적인 방법 적용: 부상확률이 높을수록 overall를 감소시키는 방향으로 실행

```python
# 부상 확률 데이터프레임과 overall 데이터프레임 병합
adjusted_df = pd.merge(df, injury_df, on='player')

# Adjusted Overall 계산
# adjusted_df['Adjusted Overall'] = adjusted_df['overall'] * (1 - np.sqrt(adjusted_df['Injury_probability']))
# 로그 스케일링 
# adjusted_df['Adjusted Overall'] = adjusted_df['overall'] * (1 - np.log1p(adjusted_df['Injury_probability']))

# 부상 확률 조정 (양수와 음수에 대해 각각 처리)
adjusted_df['Adjusted Overall'] = adjusted_df['overall'] * (1 - np.log1p(adjusted_df['Injury_probability']))

# 음수 Overall인 경우, 절대값이 증가하도록 조정
adjusted_df.loc[adjusted_df['overall'] < 0, 'Adjusted Overall'] = (
    adjusted_df['overall'] * (1 + np.log1p(adjusted_df['Injury_probability']))
)


# 결과 확인
adjusted_df.head()
# adjusted_df[['player','Adjusted Overall']]
adjusted_df.rename(columns={'player': 'PLAYER'}, inplace=True)
```
